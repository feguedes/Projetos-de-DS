### **Q4. Qual a correlação existente entre as variáveis?**

Aqui iremos limpar os dados baseados no nosso histograma, onde aparece claramente a presença de outliers.

```
df[['price','minimum_nights','number_of_reviews','calculated_host_listings_count','number_of_reviews_ltm']].describe()
```

Na tabela acima é possivel observar que na coluna `price` o valor máximo aparece de CAD $9.999.00 e na coluna `minimum_nights` apresenta a quantidade mínima de 399 dias, esses outliers prejudicam nossa análise.

Será estipulada a quantidade mínima de noites aceitável abaixo de 30 dias, assim podemos fazer uma análise mais fiel à realidade.

```
df.minimum_nights.plot(kind='box', vert=False,figsize=(15,4))
plt.show()

print('Entradas com o mínimo de noites maior que 30'.format(len(df[df.minimum_nights >30])))
print('{:.4f}%'.format((len(df[df.minimum_nights > 30]) / df.shape[0])*100))
```

Os anúncios com o mínimo de noites fora do padrão estipulado de 30 dias é de 7%, assim não prejudicando nosso `dataset` e melhorando a precisão.

Faremos a mesma coisa com a coluna `price` e será definido um IQR para estipular qual o valor aceitável sobre o mínimo de noites e preço cobrado.

```
q1_price = df.price.quantile (0.25)
q3_price = df.price.quantile (0.75)
IQR_price = q3_price - q1_price
sup_price = q3_price + 1.5 * IQR_price
inf_price = q1_price - 1.5 * IQR_price

print('IQR price', IQR_price)
print('Limite superior:' , sup_price)
print('Limite inferior: ', inf_price)
```

```
q1_minimum_nights = df.minimum_nights.quantile (0.25)
q3_minimum_nights = df.minimum_nights.quantile (0.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights

sup_minimum_nights = q3_minimum_nights + 1.5* IQR_minimum_nights
inf_minimum_nights = q1_minimum_nights - 1.5* IQR_minimum_nights
print('IQR noites mínimas:', IQR_minimum_nights)
print('Limite superior noites mínimas',sup_minimum_nights)
print('Limite inferior noites mínimas',inf_minimum_nights)
```

```
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 360.0].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.price == 0.0].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 72].index, axis=0,inplace=True)
print(df_clean.shape)
print(df.shape)
```

```
fig, ax = plt.subplots(figsize=(15,3))
df_clean.price.plot(kind='box',vert=False);
ax.set_title('Data base limpo - price')
plt.show()
print('Tamanho original: {} colunas'.format(df.shape[0]))
print('Tamanho Atual:{} colunas'.format(df_clean.shape[0]))
```

```
df_clean.describe()
```

```
# criar uma matriz de correlação
corr = df_clean[['price','minimum_nights','number_of_reviews','number_of_reviews_ltm','calculated_host_listings_count','availability_365']].corr()
display(corr)

sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='White', annot=True)
plt.show();


# mostrar a matriz de correlação
```

Comparação preço por bairro.

```
sort_price = df_clean.loc[(df_clean.price <=360) & (df_clean.price >0)].groupby('neighbourhood')['price'].median().sort_values(ascending=False).index
sns.boxplot(y='price', x='neighbourhood', data=df_clean.loc[(df_clean.price <=360) & (df_clean.price >0)],order=sort_price)
ax = plt.gca()
ax.set_xticklabels(ax.get_xticklabels(),rotation=45, ha='right')
plt.show();
```

Sem os outliers podemos dividir os bairros e fazer uma média de valor por acomodação.
A média por bairro também é possível ser análisada, assim verificando quais os bairros mais caros e os mais baratos também.

```
# ver preços por bairros, na média
df_clean.groupby('neighbourhood')['price'].median().sort_values(ascending=False)
```
