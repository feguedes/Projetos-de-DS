**Dicionário de variáveis**

* `iso_code`- Códigos de três-letras de país definido em ISO 3166-1, parte do padrão ISO 3166 publicado pela Organização Internacional para Padronização, para representar países, territórios dependentes, e zonas especiais de interesse geográfico
* `continent`- Continente geográfico
* `location`- Local geográfico
* `date`- Data da observação
* `total_cases`- Total de casos confirmados de COVID-19, pode incluir casos prováveis.
* `new_cases`- Total de novos casos, pode incluir casos prováveis.
* `new_cases_smoothed`- total de novos casos (média móvel 7 dias.
* `total_deaths`- total de mortes, pode incluir casos prováveis.
* `new_deaths`- Novas mortes atribuidas ao COVID-19, pode incluir casos prováveis.
* `new_deaths_smoothed`-Novas mortes (média móvel 7 dias) de COVID-19, pode incluir casos prováveis
* `total_cases_per_milion`-Total de casos de COVID-19 confirmados a cada 1.000.000 de pessoas, pode incluir casos prováveis.
* `new_cases_per_milion`- Novos casos de COVID-19 a cada 1.000.000 de pessoas, pode incluir casos prováveis.
* `new_cases_smoothed_per_milion`- Novos casos de COVID-19 a cada 1.000.000 de pessoas (média móvel 7 dias), pode incluir casos prováveis.
* `total_deaths_per_milion`-Total de mortes a cada 1.000.000 de pessoas.
* `new_deaths_per_milion`-Novas mortes a cada 1.000.000 de pessoas.
* `new_deaths_smoothed_per_milion`-Novas mortes (média móvel 7 dias) a cada 1.000.000 de pessoas.
* `reproduction_rate`-Estimativa em tempo real da efetividada de reprodução do COVID- 19
* `icu_patients`- Número de pacientes da COVID-19 em UTIs atualmente.
* `icu_patients_per_milion`-Número de pacientes da COVID-19 em UTIs a cada 1.000.000 de pessoas atualmente.
* `hosp_patients`-Número de pacientes da COVID-19 em hospitais.
* `hosp_patients_per_milion`-Número de pacientes da COVID-19 a cada 1.000.000 de pessoas.
* `weekly_icu_admissions`-Número de entradas nas UTIs de pacientes da COVID-19 (ultimos 6 dias)
* `weekly_icu_admissions_per_milion`-Número de entradas nas UTIs de pacientes da COVID-19 a cada 1.000.000 de pessoas (ultimos 6 dias).
* `weekly_hosp_admissions`-Número de entradas de pacientes da COVID-19 em hospitais.
* `weekly_hosp_admissions_per_milion`-Número de entradas de pacientes da COVID-19 em hospitais a cada 1.000.000.
* `total_tests`-Total de testes de COVID-19.
* `new_tests`-Novos testes de COVID-19.
* `total_tests_per_thousand`- Total de testes de COVID-19 a cada 1.000 pessoas.
* `new_tests_per_thousand`-Novos testes a cada 1.000 pessoas.
* `new_tests_smoothed`-Novos testes de COVID-19 (média móvel de 7 dias)
* `new_tests_smoothed_per_thousand`-Novos testes de COVID-19 (média móvel de 7 dias) a cada 1.000 pessoas
* `positive_rate`- Compartilhamento de testes confirmados de COVID-19 em uma média móvel de 7 dias(inverso ao `tests_per_case`)
* `test_per_case`-Testes conduzidos por novos casos confirmados de COVID-19 em uma média móvel de 7 dias (inverso ao `positive_rate`)
* `test_units`-Localização das unidades usadas para os testes.
* `total_vaccinations`- Total de doses de vacinas administradas.
* `people_vaccinated`- Número de pessoas que receberam a primeira dose da vacina.
* `people_fully_vaccinated`-Número de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação.
* `total_boosters`-Número total de doses booster administradas (essas doses vão alem do número de vacinas prescritas pelo protocolo de vacinação).
* `new_vacccinations`- Novas doses de vacinação administradas (calculada apenas por dias consecutivos).
* `new_vaccionations_smoothed`-Novas doses de vacinação (média móvel de 7  dias)
* `total_vaccinations_per_hundred`-Total de vacinas aplicadas a cada 100 pessoas da população total.
* `people_vaccinated_per_hundred`- Total de pessoas que receberam pelo menos a primeira dose da vacina a cada 100 pessoas da população total.
* `people_fully_vaccinated_per_hundred`- Total de pessoas de receberam todas as doses prescritas pelo protocolo de vacina a cada 100 pessoas da população total.
* `total_boosters_per_hundred`- Total de pessoas que receberam a dose booster administrada a cada 100 pessoas da população total.
* `new_vaccinations_smoothed_per_milion`-Total de doses de vacinação a cada 1.000.000 de pessoas (média móvel) da população total.
* `new_people_vaccinated_smoothed`-Número diario de pessoas que recebem a primeira dose de vacinação (média móvel)
* `new_people_vaccinated_smoothed_per_hundred`-Número diario de pessoas que recebem a primeira dose de vacinação (média móvel) a cada 100 pessoas da população total.
* `stringence_index`-Medidas de rigidez do gorverno: composto em mensurar 9 indicadores incluindo fechamento de escolas, locais de trabalho e fronteiras, escala de 0 a 100(100 = maior nivel de rigidez).
* `population`-População(ultimos valores disponiveis)
* `population_density`-Número de pessoas divididas por terras, mensurado em kilometros quadrados, ano mais recente díspovivel.
* `median_age`- Média de idade da população(Naçôes unidade 2020)
* `aged_65_older`-Parcela da população com 65 anos de idade, ano mais recente dísponivel.
* `aged_70_older`-Parcela da população com 70 anos de idade, ano mais recente dísponivel.
* `gdp_per_capita`- PIB per capita.
* `extreme_poverty`-Parte da população que vive em extrema pobreza (ano mais recente diponivel desde 2010).
* `cardivasc_deaths_rate`-Mortes causadas por doenças cardiovaculares in 2017(número anual de mortes a cada 100.000 pessoas)
* `diabetes_prevalence`- Incidencia de diabetes (% da população com idade entre 20 e 79 anos) em 2017
* `female_smokeres`-Parcela de mulheres fumantes, ultimo ano dísponivel.
* `male_smokers`-Parcela de homens fumantes, ultimo ano disponível.
* `handwashing_facilities`-Parcela da população com acesso a lugares para lavar as mãos, ultimo ano diponível.
* `hospital_beds_per_thousand`- Cama de hospitais disponível a cada 1.000 pessoas, ultimo ano disponível desde 2010.
* `life_expectancy`-Expectativa de vida com nascimento em 2019.
* `human_development_index`- Indicador da média das 3 dimensões básicas de desenvolvimento humano- vida longa e saudável, educação e padrão de vida descente, valores de 2019. 
* `excess_mortality`- Diferença em porcentagem entre o número de mortes reportadas semanalmente e mensalmente 2020-2021 comparado com o número de mortes projetado pelo mesmo periodo com base nos anos passados.
* `excess_mortality_cumulative`- Diferença em porcentagem entre o número de acumulo de mortes desde 1 de janeiro 2020 com o número de acumulo de mortes esperado pelos anos passados.
* `excess_mortality_cumulative_absolute`-Diferença entre o acumulo de mortes registradas desde 1 de janeiro de 2020 com a projeção de mortes esperadas pelo mesmo periodo.
* `excess_mortality_cumulative_per_milion`-Diferença entre o acumulo de mortes registradas desde 1 de janeiro de 2020 com a projeção de mortes esperadas pelo mesmo periodo a cada 1.000.000 de pessoas.
