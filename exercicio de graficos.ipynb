```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv"
df = pd.read_csv(data, index_col="data", parse_dates=True)
df.head()
```

```python
fig, ax = plt.subplots()

df["casos-confirmados"].plot(kind="line", ax=ax)
ax.set_title("Casos de Dengue Campinas")
ax.set_xlabel("Casos Confirmados")
ax.set_ylabel("Data")

plt.tight_layout()

```

```python
fig, ax = plt.subplots()


df.loc["1999-01-01":'2010-01-01', 'temperatura-maxima'].plot(kind ='line', ax=ax)
ax.set_title("Temperatura maxima")
ax.set_xlabel("De 1999 ate 2010")
ax.set_ylabel('Temperaura maxima')

plt.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows= 2, ncols=1, figsize=(10,4))
df.loc['1999-01-01' : '2008-01-01', 'temperatura-maxima'].plot(kind="line", ax=ax1)
ax1.set_title("Temperatura maxima")
ax1.set_xlabel("1999 a 2008")
ax1.set_ylabel("temperatura maxima")

df.loc['1999-01-01' : '2008-01-01', 'casos-confirmados'].plot(kind="line", ax=ax2)
ax2.set_title("Casos confirmados entre  1999 a 2008")
ax2.set_xlabel("Quantidade de casos confirmados")
ax2.set_ylabel("Periodo entre  1999 a 2008")

plt.tight_layout()
```

```python
df_norm = (df - df.mean()) / df.std()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,4))

df_norm.loc['1999-01-01' : '2008-01-01', 'temperatura-maxima'].plot(kind="line", ax=ax)

df_norm.loc['1999-01-01' : '2008-01-01', 'casos-confirmados'].plot(kind='line', ax=ax)

ax.set_title("Relacao temperatura maxima com casos confirmados")
ax.set_xlabel("Periodo de  1999 a 2008")

plt.tight_layout()
```
